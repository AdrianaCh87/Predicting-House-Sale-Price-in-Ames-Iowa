# Predicting House Sale Prices in Ames, Iowa using a Regression Model

### Exploratory Data Analysis

```python
import pandas as pd
import numpy as np

%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('tableau-colorblind10')
```

```python
path = '../datasets/'
file_name = 'cleaned_train.csv'
houses = pd.read_csv(path + file_name)
```

Review the distrubutions of the numerical features and check for normality:

```python
houses.hist(figsize=(15,15), xlabelsize=12, ylabelsize=12, layout=(7, 4))
plt.tight_layout();
```

There are a few features that are right skewed. At this time, only Sale Price will be log-transformed for the regression analysis.

Evaluate the relationship of the following features with the target feature Sale Price:

```python
sns.set(font_scale = 1)
sns.pairplot(houses,
             x_vars=['Total Bsmt SF', 'Garage Area', 'Gr Liv Area'],
             y_vars=['SalePrice'], height = 4);
plt.tight_layout();
```

```python
sns.pairplot(houses, 
             x_vars=['Overall Qual', 'TotRms AbvGrd','Garage Cars',], 
             y_vars=['SalePrice'], height = 4);
```

Positive linear relationship for these house features and Sale Price:

### Visualizations

The distribution of the houses size in Ames, IA:

```python
plt.hist(houses['Gr Liv Area'])
plt.xlabel('House Square Footage')
plt.ylabel('Number of Houses')
plt.title('Distributions of House Size in Ames, IA');
```

Investigate the number of houses sold and their average selling price in each Ames neighborhood:

```python
price_by_place = houses.groupby('Neighborhood')['SalePrice'].mean()
houses_by_place = houses.groupby('Neighborhood')['Neighborhood'].count()
neighborhoods = price_by_place.index

# Sort by number of houses sold in each neighborhood
idx = houses_by_place.argsort()
neighborhoods,houses_by_place,price_by_place=[np.take(x, idx) for x in [neighborhoods,houses_by_place,price_by_place]]

y = np.arange(price_by_place.size)

fig, axes = plt.subplots(ncols=2, sharey=True, figsize=(12,8))
axes[0].barh(y, houses_by_place, align='center')
axes[0].set(title='Number of Houses Sold')
axes[1].barh(y, price_by_place, align='center')
axes[1].set(title='Average House Price in $')

axes[0].invert_xaxis()
axes[0].set(yticks=y, yticklabels=neighborhoods)
axes[0].yaxis.tick_right()

for ax in axes.flat:
    ax.margins(0.08)
    ax.grid(True)

fig.tight_layout()
fig.subplots_adjust(wspace=0.15)
plt.show()
```

```python
plt.figure(figsize=(12,6))
houses.groupby(['Yr Sold', 'Mo Sold'])['SalePrice'].count().plot.bar()
plt.xlabel('Month and Year when House Sold')
plt.ylabel('Number of Houses Sold');
```

This graph shows a cyclical pattern where the number of houses sold peaks in the summer months, and is low at the end/beginning of the year

```python
plt.figure(figsize = (8, 5))
yrbuilt = houses.groupby('Year Built')['SalePrice'].mean()
yrbuilt.sort_index().plot.line();
plt.ylabel('Avg House Price $')
plt.title('Year when House was Built and its Sale Price');
```

```python
plt.figure(figsize=(8,5))
houses.groupby('Fireplaces')['SalePrice'].mean().sort_values().plot.bar()
plt.ylabel('Avg House Sale Price $')
plt.title('The Average House Price for the Number of Fireplaces in a House');
```

```python
plt.figure(figsize = (8, 5))
houses.groupby('TotRms AbvGrd')['SalePrice'].median().plot.bar();
plt.ylabel('Median House Price $')
plt.xlabel('Number of Rooms')
plt.title('Number of rooms and Sale Price');
```
